```python
import numpy as np
import matplotlib.pyplot as plt
import random
from sympy.utilities.iterables import multiset_permutations
```

```python
def Qiskit_Random_Gen(N):
    import qiskit
    import numpy as np
    sim = qiskit.Aer.get_backend('qasm_simulator')
    n = int(np.log2(N))
    # print(n)
    circ = qiskit.QuantumCircuit(n)
    circ.h(list(range(n)))
    circ.measure_all()

    t_qc = qiskit.transpile(circ, sim)
    counts = sim.run(t_qc,shots=1).result().get_counts()

    result = int(list(counts.keys())[0],2)
    
    return result
   
    
```

```python

def randomList_Quantum(m, n):
 
    # Create an array of size m where
    # every element is initialized to 0
    arr = [0] * m;
     
    # To make the sum of the final list as n
    for i in range(n) :
 
        # Increment any random element
        # from the array by 1
        arr[Qiskit_Random_Gen(n) % m] += 1
    return arr

def Update(E,nop):
 
    Enew = randomList_Quantum(nop, E)
    sm = np.sum(Enew)

    return Enew

def Entropy(E,Es):
    ps = list(multiset_permutations(Es))
    microstates = 0
    for p in ps:
        sm = np.sum(p)
        # print(sm,p,E)
        if sm == E:
            # print(p,E)
            microstates+=1
    entropy = np.log(microstates)
    return entropy


    
    
```
